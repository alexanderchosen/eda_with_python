### Data Preparation

This indicates that the dataset has 493 rows and 16 columns. It has no null values, and all the columns are needed for this analysis. However, I need to convert some of the columns to numerical data or a sort of grade. Examples are the Attendance, preparation and gaming

```python
# to convert preparation column to numeric grades
# that is, 1 for 0-1 hour, 2 for 2-3 hours, 3 for more than 3 hours
data['Preparation'].value_counts()
```

```python
data['Preparation'].unique()
```

```python
# define mapping using dictionary
preparation_dict = {
    "0-1 Hour": 1,
    "2-3 Hours": 2,
    "More than 3 Hours": 3
}

preparation_dict
```

```python
# use the dictionary to create a new column
data['preparation_level'] = data['Preparation'].map(preparation_dict)
data.head(20)
```

```python
data[['Preparation', 'preparation_level']].head()
```

```python
data['preparation_level'].value_counts()
```

```python
# to convert the gaming column data from a set of range values to a numeric value representing each range
data['Gaming'].unique()
```

```python
# next, after checking its unique values, create a gaming dictionary
gaming_dict = {
    "0-1 Hour":1,
    "2-3 Hours": 2,
    "More than 3 Hours": 3
} 

gaming_dict
```

```python
# next, map the dictionary values to the new column
data['gaming_level'] = data['Gaming'].map(gaming_dict)
data.head(20)
```

```python
# Finally, to convert the Attendance column into Attendance levels
# we check for the unique values first
data['Attendance'].unique()
```

```python
# next, i will create a dictionary
attendance_dict ={
    "80%-100%": 4,
    "60%-79%": 3,
    "40%-59%": 2,
    "Below 40%": 1
}
attendance_dict
```

```python
# finally, map the dictionary to the new column 
data['attendance_level']= data['Attendance'].map(attendance_dict)
data.head(20)
```
