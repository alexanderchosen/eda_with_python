## Data Preparation

```python
df = data[['coaster_name', 
      #'Length', 'Speed', 
      'Location', 'Status', 
      #'Opening date','Type', 
      'Manufacturer', 
      #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
      # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
       'latitude', 'longitude', 
       'Type_Main',
       'opening_date_clean', 'speed1', 
       # 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
      # 'height_value', 'height_unit',
         'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
# drop a single column
df = df.drop(['speed1'], axis = 1)
```

```python
df.shape
```

```python
df.dtypes
```

```python
# to convert the opening_date_clean column to date type
df['opening_date_clean']= pd.to_datetime(df['opening_date_clean'])

# to convert to a numeric data:
#df['column_name'] = pd.to_numeric(['column_name'])
```

```python
df.dtypes
```

```python
df.head
```

```python
# renaming columns
df.columns
df = df.rename(columns={'Type_Main': 'main_type',
                        'opening_date_clean': 'opening_date',
                        'Inversions_clean': 'inversions',
                        'Gforce_clean': 'Gforce'})
```

```python
df.head
```

```python
# check for missing or null values
df.isna().sum()
```

```python
df.duplicated()
```

```python
# to check if multiple columns has duplicates
df.loc[df.duplicated(subset = ['coaster_name', 'Location', 'year_introduced'])]
```

```python
# to check if a specific column is duplicated
df.loc[df.duplicated(subset = ['coaster_name'])]
```

```python
df.query('coaster_name == "Crystal Beach Cyclone"')
```

```python
# to remove duplicated columns, we use the ~ sign to take the inverse of the duplicated data which are the originals
# to also ensure that the indexing of the rows reset since we are removing duplicated rows, we use the .reset_index(drop= True)

df = df.loc[~df.duplicated(subset = ['coaster_name', 'Location', 'opening_date'])].reset_index(drop = True).copy()
```

```python
# Answering Questions From Your Data

# 1 - which of the location has the highest coaster speed and has more than 10 coasters available

df['Location'].value_counts()
```

```python
ax = df.query('Location != "Other"').groupby('Location')['speed_mph'] \
    .agg(['mean', 'count']).query('count >= 10') \
    .sort_values('mean')['mean'].plot(kind = 'barh', figsize = (12, 5), title = 'Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')

plt.show()
```
